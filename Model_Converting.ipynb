```python
import torchvision.models as models
import torch
import torch.onnx

# tải mô hình đã huấn luyện của torchvision
resnet50 = models.resnet50(pretrained=True, progress=False).eval()
```

```python
# Định nghĩa dữ liệu mẫu để convert sang onnx mô hình 
# có 2 loại kiểu batching . 1 là tĩnh batching, 2 là động batching 
# Đây là ví dụ về tĩnh batching
BATCH_SIZE=4
dummy_input=torch.randn(BATCH_SIZE, 3, 224, 224)
torch.onnx.export(resnet50, dummy_input, "resnet50_pytorch.onnx", verbose=False)
```

```python
# tải ảnh mẫu và hiển thị
from skimage import io
from skimage.transform import resize
from matplotlib import pyplot as plt
import numpy as np

url='https://images.dog.ceo/breeds/retriever-golden/n02099601_3004.jpg'
img = resize(io.imread(url), (224, 224)) # thay đổi kích thươc (rộng,dài)
plt.imshow(img)
```

```python
import torch
from torchvision.transforms import Normalize

# tạo tiền xử lí cho ảnh trước khi qua onnx model
def preprocess_image(img):
    norm = Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    result = norm(torch.from_numpy(img).transpose(0,2).transpose(1,2))
    return np.array(result, dtype=np.float32)
```

```python
#load model onnx
import onnxruntime
ort_session = onnxruntime.InferenceSession("resnet50_pytorch.onnx")
```

```python
# Do là mình đang dùng tĩnh batching nên cần 4 ảnh, giả sử có 4 ảnh
imgs = [img,img,img,img]
image_preprocessed = np.array([preprocess_image(img) for img in imgs])
```

```python
ort_inputs = {ort_session.get_inputs()[0].name: image_preprocessed}
ort_outs = ort_session.run(None, ort_inputs)
```
